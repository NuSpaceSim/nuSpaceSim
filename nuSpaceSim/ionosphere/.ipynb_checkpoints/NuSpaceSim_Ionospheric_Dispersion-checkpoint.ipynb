# Ionospheric Dispersion and Dedispersion

This notebook includes the functions to model ionospheric dispersion and approaches to dedispersion.

The model of ionospheric dispersion follows the treatment of Moses & Jacobson (2004). We recap the key ideas here.

For a plane wave propagating in a medium with index of refreaction $n(f, s)$ that depends on frequency $f$ and position $s$ along the line of propagation, the phase at the receiver is given by
$$\phi(f)=\frac{2\pi}{c}\int_{T_X}^{R_X} \ ds \ f \ n(f,s).$$

The frequency-depenent delay is $$\tau(f)= \frac{1}{2\pi}\frac{\partial\phi}{\partial f}.$$

The index of refraction is given by 
$$n^2 = 1-X \left[1-\frac{Y^2 \sin^2\beta}{2\left(1-X\right)}\pm \sqrt{\frac{Y^4\sin^4\beta}{4(1-X^2)}+Y^2\cos^2\beta}\right]^{-1}$$
where $X=(f_{pe}/f)^2$, $Y=f_{ce}/f$, $f_{pe}$ is the electron plasma frequency, $f_{ce}$ is the cyclotron frequency and $\beta$ is the angle between the geomagnetic field and the direction of wave propagation. 

While in principle one could obtain the dispersion delay by numerical integration with the expression above, the results do depend on the electron density profile of the ionosphere as well spatial variations in the Earth's geomagnetic field. Moses & Jacobson 2004 uses an approximation up to fourth order in $f^{-n}$ to obtain
$$\tau(f)=h/c+C_2f^{-2} + s C_3f^{-3} + C_4f^{-4}.$$ $C_2$ depends only on the total electron content (TEC) along the line of sight while $C_3$ depdends on TEC and the geomagnetic field. $C_4$ depends on the altitude distribution of electrons as do higher order terms in the expansion.

In this notebook we deal only with the approximation. It would be a useful exercise in the future to compare the expansion to a numerical integration along the line of sight using instances of the International Reference Ionosphere and the International Reference Geomagnetic Field models.

**References** 

Moses & Jacobson (2004), "Ionospheric profiling through radio-frequency signals recorded by the FORTÃ‰ satellite, with comparison to the International Reference Ionosphere", Advances in Space Research, 34, 2096.

International Reference Ionosphere: https://iri.gsfc.nasa.gov/

### Electromagnetic Impulse Waveform Models

For this study we use a simple input waveform and noise model. The waveform is a band-limited delta function in a user-specified band spanning frequencies $f_1$ and $f_2$. The user can specify the delay, which in this case simply means the location of the peak. The user also specifies the record length, which will be import later for dispersion. The waveform model takes a spectral index $k$ as an input so that the pulse spectrum is proportional to $f^k$. For more advanced studies, we will want to use the exponentially falling spectrum of radio emission from extensive air showers or, better yet, use ZHAireS-simulated pulses as an input. The peak of the pulse is normalized to unity so that the user can later specify the scale of the amplitude. 

The noise model coded below uses a Rician distribution and is band-limited by a user-specified band spanning frequencies $f_1$ and $f_2$. The noise model takes a spectral index $k$ as an input so that the noise spectrum is proportional to $f^k$. The noise RMS is normalized to unity so that the user can specify the relative amplitude separately. 

```python


def wfm(f1,f2, fs,delay, record_length,index):
    dt=1./fs
    #print dt
    N=int(record_length/dt)
    if(N%2!=0): N+=1
    df=fs/float(N)
    w=arange(N,dtype=float64)
    f=df*arange(N,dtype=float64)
    fft_w=np.fft.rfft(w)*0.
    i_f1=int(f1/df)
    i_f2=int(f2/df)

    fft_w[i_f1:i_f2]=(f[i_f1:i_f2])**index
    fft_w[i_f1:i_f2]*=exp(-1j*2.*pi*delay*f[i_f1:i_f2])
    w=np.fft.irfft(fft_w)
    norm = np.max(np.abs(w))

    w/=norm
    f=df*arange(N,dtype=float64)
    t=dt*arange(N,dtype=float64)


    return t,w

def noise(f1,f2, fs,record_length,index):
    #fs=2.*f2
    #print f1,f2,fs
    dt=1./fs
    #print dt
    N=int(record_length/dt)
    if(N%2!=0): N+=1
    df=fs/N
    w=0.*arange(N,dtype=float64)
    n_fft=np.fft.rfft(w)
    f =np.arange(len(n_fft))/fs
    r_re = np.random.normal(0.,1., len(n_fft))
    r_im = np.random.normal(0.,1., len(n_fft))
    i_f1=int(f1/df)
    i_f2=int(f2/df)
    n_fft[1:]=(r_re[1:] + 1j*r_im[1:])*(f[1:]*1.e-6)**index
    n_fft[0:i_f1]*=0.
    n_fft[i_f2:len(f)-1]*=0.

    #(random,gauss(0.,1)+1j*random,gauss(0.,1))
    w_noise=real(np.fft.irfft(n_fft))
    w_noise/=sqrt(cumsum(w_noise**2)[len(w)-1]/len(w))
    return w_noise

```

```python
'''
Example of an impulsive waveform
'''

f1=30.e6
f2=1000.e6
fs=2000.e6
record_length = 4096./fs
delay=3.e-6
#print (C2,C3, C4)
sig_index = -1.
t1, w1 = wfm(f1, f2, fs, delay, record_length,sig_index)
figure()
plot(t1*1.e6, w1, 'k', alpha=1.)
print('len(w1)', len(w1))
xlabel('Time, ns')
ylabel('Amplitude, arb. u.')

'''
Zoom in
'''
figure()
plot(t1*1.e6, w1, 'k', alpha=1.)
ipk = np.argmax(w1)
xlim(t1[ipk]*1.e6-0.25, t1[ipk]*1.e6 + 0.25)
xlabel('time, ns')
ylabel('Amplitude, arb. u.')
```

```python
def dispersion_phases(frq, TEC, Bcos_theta):
    alpha=1.3445e9 # (Hz/TECU)
    em_ratio=1.759e11 # Hz / Tesla, this is the electron charge to mass ratio
    c = 2.99e8
    #Bcos_theta=1.e-6 # Tesla
    C2=alpha*TEC
    C3=alpha*TEC*2.*em_ratio*Bcos_theta#/(2.*pi)
    C4a =  (3./2.)* alpha**2 * TEC**2 * c/(250.e3) # 250 km is the effective slab thickness of the ionosphere.
    C4b = 3./2. * C2 * 1.5*2.*(Bcos_theta*em_ratio)**2 
    #print('sqrt(C2) MHz', C2*1.e-6)
    #print('C3/C2 MHz', C3/C2*1.e-6)
    #print('sqrt(C4a/C2) MHz', sqrt(C4a/C2)*1.e-6)
    #print('sqrt(C4b/C2)MHz', sqrt(C4b/C2)*1.e-6)
    phase2 = np.zeros(len(frq))
    phase3 = np.zeros(len(frq))
    phase4a = np.zeros(len(frq))
    phase4b = np.zeros(len(frq))
    phase2[1:]=C2/(frq[1:])
    phase3[1:]=C3/(frq[1:]**2)
    phase4a[1:]=C4a/(frq[1:]**3)
    phase4b[1:]=C4b/(frq[1:]**3)
    #print ('phase2', phase2)
    #print ('phase3', phase3)
    #print ('phase4', phase4)
    #print ('30 MHz: %1.2e %1.2e %1.2e'%(C2/30.e6, C3/(30.e6)**2, C4/(30.e6)**3))
    return phase2, phase3, phase4a, phase4b

def disperse(wfm_h, wfm_v, fs, TEC, Bcos_theta):
    '''
    #TEC=10.
    alpha=1.3445e9
    em_ratio=1.759e11 # Hz / Tesla, this is the electron charge to mass ratio
    c = 2.99e8
    #Bcos_theta=1.e-6 # Tesla
    C2=alpha*TEC
    C3=alpha*TEC*2.*em_ratio*Bcos_theta
    C4=0.
    '''
    fft_w_h=np.fft.rfft(wfm_h)
    fft_w_v=np.fft.rfft(wfm_v)
    
    fft_w_x=(fft_w_h-1j*fft_w_v)/np.sqrt(2.)
    fft_w_o=(fft_w_h+1j*fft_w_v)/np.sqrt(2.)
    
    N=len(fft_w_x)
    df=fs/N
    f=df*arange(N,dtype=float64)
    #for k in range(0,N/2):
    '''
    phase2 = np.zeros(len(f))
    phase3 = np.zeros(len(f))
    phase4 = np.zeros(len(f))
    phase2[1:]=C2/f[1:]
    phase3[1:]=C3/f[1:]**2
    phase4[1:]=C4/f[1:]**3
    '''
    
    phase2, phase3, phase4a, phase4b = dispersion_phases(f, TEC, Bcos_theta)
    
    # yes, it needs the factor of 2*pi below
    
    fft_w_x[1:]*=exp(1j*2.*pi*phase2[1:])
    fft_w_x[1:]*=exp(+1j*2.*pi*phase3[1:])
    fft_w_x[1:]*=exp(1j*2.*pi*(phase4a[1:]+phase4b[1:]))

    fft_w_o[1:]*=exp(1j*2.*pi*phase2[1:])
    fft_w_o[1:]*=exp(-1j*2.*pi*phase3[1:])
    fft_w_o[1:]*=exp(1j*2.*pi*(phase4a[1:]+phase4b[1:]))

    
    fft_w_h=   (fft_w_x + fft_w_o)/np.sqrt(2.)
    fft_w_v=1j*(fft_w_x - fft_w_o)/np.sqrt(2.)

    #fft_w_h = fft_w_x.copy()
    #fft_w_v = fft_w_o.copy()
    t=arange(len(wfm_h),dtype=float64)/fs
    #figure(1)
    #plot(f[1:N/2],-diff(phase[0:N/2])/diff(f[0:N/2]))
    #xlim(30.e6,300.e6)
    #print(f)
    ##plot(t*1.e6,ifft(fft_w))
    #specgram(ifft(fft_w_h))
    #show()
    return t, np.fft.irfft(fft_w_h), np.fft.irfft(fft_w_v)
```

For a waveform 
$$|{\psi}\rangle=\cos\theta |h\rangle + \sin\theta|v\rangle$$
the ionosphere disperses signals according to
$$|{\psi_d}\rangle=|x\rangle\langle x|\psi\rangle e^{i\alpha_x}  +|o\rangle\langle o|\psi\rangle e^{i\alpha_o}$$
At the detector
$$\langle h|{\psi_d}\rangle=\langle h|x\rangle\langle x|\psi\rangle e^{i\alpha_x}  +\langle h|o\rangle\langle o|\psi\rangle e^{i\alpha_o}$$
$$\langle v|{\psi_d}\rangle=\langle v|x\rangle\langle x|\psi\rangle e^{i\alpha_x}  +\langle v|o\rangle\langle o|\psi\rangle e^{i\alpha_o}$$

Using the relations

$|x\rangle = (|h\rangle - i|v\rangle)/\sqrt{2}$

$|o\rangle = (|h\rangle + i|v\rangle)/\sqrt{2}$

$|h\rangle = (|x\rangle + |o\rangle)/\sqrt{2}$

$|v\rangle = i(|x\rangle - |o\rangle)/\sqrt{2}$

We have

$\langle x|\psi\rangle = e^{i\theta}/\sqrt{2}$

$\langle o|\psi\rangle = e^{-i\theta}/\sqrt{2}$

$$\langle h|{\psi_d}\rangle=
\frac{e^{i(\alpha_o-\theta)}+e^{i(\alpha_x+\theta)}}{2}$$
$$\langle v|{\psi_d}\rangle=i\frac{e^{i(\alpha_o-\theta)}-e^{i(\alpha_x+\theta)}}{2}$$

So with one polarization alone, we have

$$\langle h|{\psi_d}\rangle e^{-i\alpha_o}=
\frac{e^{-i\theta}+e^{i(\alpha_x-\alpha_o+\theta)}}{2}$$
$$\langle h|{\psi_d}\rangle e^{-i\alpha_x}=
\frac{e^{i(\alpha_o-\alpha_x-\theta)}+e^{i\theta}}{2}$$
$$\langle v|{\psi_d}\rangle e^{-i\alpha_o}=
i\frac{e^{-i\theta}-e^{i(\alpha_x-\alpha_o+\theta)}}{2}$$
$$\langle v|{\psi_d}\rangle e^{-i\alpha_x}=
i\frac{e^{i(\alpha_o-\alpha_x-\theta)}-e^{i\theta}}{2}$$

We want to identify the most robust combination of parameters above. Cannot rely in exquisite calibrations for the first pass.

You want to calculate the dispersion phases and then do the linear algebra.

```python
from scipy import signal

'''
Define the band and sampling frequency
'''
f1=30.e6
f2=300.e6
fs=600.e6

'''
Note: the record length has to be larger than the dispersion. 
The user needs to ensure this is the case.
'''
record_length = 2.**14/fs
delay=3.e-6
sig_index = -1.
t1, w1 = wfm(f1, f2, fs, delay, record_length,sig_index)

w2_h= 0.*w1.copy()
w2_v= w1.copy()
TEC = 10.
Bcos_theta = 50.e-6 # in Tesla

t3, w3_h, w3_v =disperse(w2_h,   w2_v, fs, TEC, Bcos_theta)

plot(t3*1.e6, w3_v, alpha=0.3)
plot(t3*1.e6, w3_h, alpha=0.3)
#xlim(delay*1.e6-0.5, delay*1.e6 + 5.)


figure()
f, t, Sxx = signal.spectrogram(w3_h, fs, nperseg=2**6)
print('np.min(Sxx)',np.min(Sxx))
print('np.max(Sxx)',np.max(Sxx))
pcolormesh(t*1.e6, f, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
ylabel('Frequency [Hz]')
xlabel('Time [$\mu$sec]')
figure()
f, t, Sxx = signal.spectrogram(w3_v, fs, nperseg=2**6)
pcolormesh(t*1.e6, f, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
ylabel('Frequency [Hz]')
xlabel('Time [$\mu$sec]')
```

```python
'''
Define the TEC search function.
This will produce the waveform with specified parameters (f1, f2, fs, record_length, delay, 
TEC, Bcos_theta, SNRa, pol_ang, sig_index, index (for noise))
and scan over a range of TECs specified by the user (dTEC, Nscan).
'''
def TEC_search(f1=30.e6, f2=80.e6, fs=200.e6, 
               record_length = 4096./200.e6, 
               delay = 3.e-6,
               TEC=10., 
               Bcos_theta = 41.9e-6,
               SNRa = 10., 
               pol_ang = -1,     
               sig_index = 1.,
               index = -2.6,
               dTEC = 0.05,
               Nscan = 101,
               verbose=True, plots=True):

    '''
    Prepare dispersed waveforms
    '''

    t1, w1 = wfm(f1, f2, fs, delay, record_length,sig_index)
    
    '''
    Use a randomly sampled linear polarization angle
    '''
    pang = pol_ang*pi/180.
    if(pol_ang==-999):
        pang = np.random.uniform(-pi/2., pi/2.)
    #pang = -12.*pi/180.

    '''
    Split the waveform into H and V pol components
    '''
    w2_h= np.cos(pang)*w1.copy()
    w2_v= np.sin(pang)*w1.copy()

    '''
    Disperse the waveform
    '''
    t3, w3_h, w3_v =disperse(w2_h,   w2_v, fs, TEC, Bcos_theta)

    '''
    Simulate a noise waveform for each polarization
    '''
    wn_h   = noise(f1, f2, fs, record_length,index)
    wn_v   = noise(f1, f2, fs, record_length,index)

    '''
    Normalize the noise RMS to unity
    '''
    wn_h/= np.std(wn_h)
    wn_v/= np.std(wn_v)

    '''
    Add the noise to the dispersed waveforms. 
    '''
    w3_h = SNRa*w3_h + wn_h
    w3_v = SNRa*w3_v + wn_v
    
    '''
    Set up TEC scan list. 
    This starts with the true value and scans by a specified number of steps and step resolution.
    '''
    tTEC_list = np.arange(TEC-Nscan/2.*dTEC, TEC+(Nscan/2+1.e-3)*dTEC, dTEC)
    t0, w0 = wfm(f1, f2, fs, -3.e-6, record_length,sig_index)


    if(plots):
        figure()
        plot(t3*1.e6, w3_h, 'b-', alpha=0.3, label='H')
        plot(t3*1.e6, w3_v, 'r-', alpha=0.3, label='V')
        #xlim(delay*1.e6-0.5, delay*1.e6 + 5.)
        xlabel('Time, [$\mu$s]')
        ylabel('Amplitude [$\sigma_\mathrm{RMS}$]')
        legend(loc=1)
        grid(True, which='both', alpha=0.2)
        tight_layout()
        savefig('dispersed_wfm.pdf')

        figure(figsize=(6,7))
        subplot(211)
        f, t, Sxx = signal.spectrogram(w3_h, 200.e6, nperseg=2**6)
        #print('np.min(Sxx)',np.min(Sxx))
        #print('np.max(Sxx)',np.max(Sxx))
        pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
        ylabel('Frequency [MHz]')
        xlabel('Time [$\mu$sec]')
        subplot(212)
        f, t, Sxx = signal.spectrogram(w3_v, 200.e6, nperseg=2**6)
        #print('np.min(Sxx)',np.min(Sxx))
        #print('np.max(Sxx)',np.max(Sxx))
        pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
        ylabel('Frequency [MHz]')
        xlabel('Time [$\mu$sec]')
        tight_layout()
        savefig('Lin_pol_specgram.pdf')

    if(plots):
        figure(22)
    if(verbose):
        print('')
        print('pang deg', pang*180./pi)
        print('')
    v1 = []
    v2 = []
    v3 = []
    '''
    Loop over TEC templates
    '''
    for tTEC in tTEC_list:
        '''
        De-dispersion template for correlation
        '''
        #tt, wt_h, wt_v = disperse(w0.copy()/np.sqrt(2.),  w0.copy()/np.sqrt(2.), fs, -tTEC, Bcos_theta)

        N=len(np.fft.rfft(w3_h))
        df=fs/N
        frq=df*arange(N,dtype=float64)

        '''
        Model the X and O mode waveforms as a function of tTEC and Bcos_theta
        '''
        ph2, ph3, ph4a, ph4b = dispersion_phases(frq, tTEC, Bcos_theta)
        fft_wX = exp(-1j*2.*pi*(ph2+ph3+ph4b))
        fft_wO = exp(-1j*2.*pi*(ph2-ph3+ph4b))
        wX = np.fft.irfft(fft_wX)
        wO = np.fft.irfft(fft_wO)
        
        '''
        Model the linearly polarized versions (labeled "P" and "M" for plus and minus)
        '''
        wP = np.fft.irfft(   (fft_wX + fft_wO))/np.sqrt(2.)
        wM = np.fft.irfft(-1j*(fft_wX - fft_wO))/np.sqrt(2.)
        '''
        Cross Correlate (or convolve) the X, O, P, and M templates with the simulated h and v waveforms.
        '''
        corr_Xh = np.convolve(wX, w3_h)
        corr_Oh = np.convolve(wO, w3_h)
        corr_Xv = np.convolve(wX, w3_v)
        corr_Ov = np.convolve(wO, w3_v)

        corr_Ph = np.convolve(wP, w3_h)
        corr_Mh = np.convolve(wM, w3_h)
        corr_Pv = np.convolve(wP, w3_v)
        corr_Mv = np.convolve(wM, w3_v)
        
        '''
        Save the incoherently summed cross-correlation peaks for X,O modes and P,M waveforms.
        '''
        v1.append(np.sqrt(np.max(corr_Xh**2+corr_Oh**2+corr_Xv**2+corr_Ov**2)))
        v2.append(np.max(np.sqrt(corr_Ph**2 + corr_Pv**2 + corr_Mh**2 + corr_Mv**2)))


        if (plots):
            if(tTEC == 10.):
                figure(18)
                f, t, Sxx = signal.spectrogram(wX, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Omode_specgram.pdf')
                figure(19)
                f, t, Sxx = signal.spectrogram(wO, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Xmode_specgram.pdf')

            figure(22)
            plot([tTEC], [np.max(np.sqrt(corr_Ph**2 + corr_Pv**2 + corr_Mh**2 + corr_Mv**2))], 'go', mfc='none')
            xlabel('Template TEC, TECU')
            ylabel('Amplitude [$\sigma_\mathrm{RMS}$]')
            grid(True, which='both', alpha=0.2)
            ylim(0., 100.)
            yticks(np.arange(0., 100.1, 10.))
            xlim(7., 13.)
            xticks(np.arange(7., 13.1, 1.))
            tight_layout()
            savefig('TEC_scan.pdf')

            '''
            Make plots near the true solution.
            '''
            if( np.abs(tTEC-10.) < 0.025):
                figure(40)
                ctf = np.logical_and(frq>30.e6, frq<80.e6)
                plot(ph2[ctf]/frq[ctf]*1.e6, frq[ctf]*1.e-6, '--', label=r'$\tau_2$')
                plot(ph3[ctf]/frq[ctf]*1.e6, frq[ctf]*1.e-6, '-.', label=r'$\tau_3$')
                plot(ph4a[ctf]/frq[ctf]*1.e6, frq[ctf]*1.e-6, ':', label=r'$\tau_{4a}$')
                plot(ph4b[ctf]/frq[ctf]*1.e6, frq[ctf]*1.e-6, ':', label=r'$\tau_{4b}$')
                plot((ph2[ctf]+ ph3[ctf]+ (ph4a[ctf]+ph4b[ctf]))/frq[ctf]*1.e6, frq[ctf]*1.e-6, lw=5, label=r'$\tau_o$')
                plot((ph2[ctf]- ph3[ctf]+ (ph4a[ctf]+ph4b[ctf]))/frq[ctf]*1.e6, frq[ctf]*1.e-6, lw=5, label=r'$\tau_x$')
                
                id1 = np.argmin(np.abs(frq-30.e6))
                id2 = np.argmin(np.abs(frq-80.e6))
                print('tst: f1 %1.2f f2 %1.2f t2 %1.2f t3 %1.2f t4a %1.3f t4b %1.3f '%(
                    frq[id1]*1.e-6, frq[id2]*1.e-6,
                    (ph2/frq*1.e6)[id1] - (ph2/frq*1.e6)[id2],
                    (ph3/frq*1.e6)[id1] - (ph3/frq*1.e6)[id2],
                    (ph4a/frq*1.e6)[id1] - (ph4a/frq*1.e6)[id2],
                    (ph4b/frq*1.e6)[id1] - (ph4b/frq*1.e6)[id2]))
                      
                xlabel('Delay [$\mu$s]')
                ylabel('Frequency [MHz]')
                legend(loc=0)
                grid(True, which='both', alpha=0.2)
                xlim(0.)
                tight_layout()
                savefig('dispersion_delays.pdf')
                #print('frq', frq[ctf])
                figure(30)
                t_corr = np.cumsum(np.ones(len(corr_Ph)))/(200.e6)
                plot(t_corr*1.e6, corr_Ph, '-', lw=2., label = '$C_{hh}$')
                plot(t_corr*1.e6, corr_Mv, '--',lw=1.5,  label = '$C_{vv}$')
                plot(t_corr*1.e6, corr_Mh, ':', lw=1.5, label = '$C_{hv}$')
                plot(t_corr*1.e6, corr_Pv, '-.',lw=1.5,  label = '$C_{vh}$')
                ylabel('Amplitude [$\sigma_\mathrm{RMS}$]')
                xlabel('Time [$\mu$sec]')
                legend(loc=2, title=r'$\theta=%1.1f^\circ$'%(pang*180./pi), fontsize=12, title_fontsize=12)
                t_pk = t_corr[np.argmax(np.abs(corr_Pv))]*1.e6
                if(np.max(np.abs(corr_Ph)) > np.max(np.abs(corr_Pv))):
                    t_pk = t_corr[np.argmax(np.abs(corr_Ph))]*1.e6
                    
                xlim(t_pk - 0.25, t_pk+0.25)
                print('tTEC', tTEC)
                grid(True, which='both', alpha=0.2)
                tight_layout()
                savefig('corr_wfms.pdf')
                
                #print(np.max(np.abs(corr_Mh)))
                figure(23)
                f, t, Sxx = signal.spectrogram(corr_Ph, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Chh_specgram.pdf')

                figure(24)
                f, t, Sxx = signal.spectrogram(corr_Mh, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Cvh_specgram.pdf')

                figure(25)
                f, t, Sxx = signal.spectrogram(corr_Pv, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Chv_specgram.pdf')

                figure(26)
                f, t, Sxx = signal.spectrogram(corr_Mv, 200.e6, nperseg=2**6)
                pcolormesh(t*1.e6, f*1.e-6, np.log10(Sxx), vmin = np.log10(np.max(Sxx))-2.)
                ylabel('Frequency [MHz]')
                xlabel('Time [$\mu$sec]')
                tight_layout()
                savefig('Cvv_specgram.pdf')

    figure()
    mx = np.max([np.max(v1), np.max(v2)])
    delta = mx/25.
    bins = np.arange(0.-delta/2., mx+delta/2., delta)
    hist(v1, bins=bins, label='X-O wfms')
    hist(v2, bins=bins, label='P-M wfms', alpha=0.5)
    legend(loc=0)
    xlabel('Peak Summed Cross-Correlation, arb. u.')
    if verbose:
        print('np.mean(v1), np.mean(v2)',np.mean(v1), np.mean(v2))
        print('\n')
        print('np.std(v1), np.std(v2): %1.3e %1.3e'%(np.std(v1), np.std(v2)))
        print('np.max(v1), np.max(v2): %1.3e %1.3e'%(np.max(v1), np.max(v2)))
        
    return np.max(v1), np.max(v2)
```
